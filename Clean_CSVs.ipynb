```python
# import tools and csvs
import pandas as pd
import numpy as np

# NOTE: The CDC website wouldn't allow zero values because the dataset would be too big.

df_2005_2006 = pd.read_csv("Raw_CSVs/2005-2006.csv")
df_2007_2008 = pd.read_csv("Raw_CSVs/2007-2008.csv")
df_2009_2010 = pd.read_csv("Raw_CSVs/2009-2010.csv")
df_2011_2012 = pd.read_csv("Raw_CSVs/2011-2012.csv")
df_2013_2014 = pd.read_csv("Raw_CSVs/2013-2014.csv")
df_2015_2016 = pd.read_csv("Raw_CSVs/2015-2016.csv")
df_2017_2018 = pd.read_csv("Raw_CSVs/2017-2018.csv")
```

```python
# remove null values and Notes column from each data set.
df_2005_2006 = df_2005_2006[df_2005_2006['Year'].notnull()]
df_2005_2006 = df_2005_2006.drop(columns=['Notes'])
df_2005_2006.head()
```

```python
df_2007_2008 = df_2007_2008[df_2007_2008['Year'].notnull()]
df_2007_2008 = df_2007_2008.drop(columns=['Notes'])
df_2007_2008.head()
```

```python
df_2009_2010 = df_2009_2010[df_2009_2010['Year'].notnull()]
df_2009_2010 = df_2009_2010.drop(columns=['Notes'])
df_2009_2010.head()
```

```python
df_2011_2012 = df_2011_2012[df_2011_2012['Year'].notnull()]
df_2011_2012 = df_2011_2012.drop(columns=['Notes'])
df_2011_2012.head()
```

```python
df_2013_2014 = df_2013_2014[df_2013_2014['Year'].notnull()]
df_2013_2014 = df_2013_2014.drop(columns=['Notes'])
df_2013_2014.head()
```

```python
df_2015_2016 = df_2015_2016[df_2015_2016['Year'].notnull()]
df_2015_2016 = df_2015_2016.drop(columns=['Notes'])
df_2015_2016.head()
```

```python
df_2017_2018 = df_2017_2018[df_2017_2018['Year'].notnull()]
df_2017_2018 = df_2017_2018.drop(columns=['Notes'])
df_2017_2018.head()
```

```python
# Export dataframes to csvs
df_2005_2006.to_csv('Clean_CSVs/clean_2005-2006.csv', index=False)
df_2007_2008.to_csv('Clean_CSVs/clean_2007_2008.csv', index=False)
df_2009_2010.to_csv('Clean_CSVs/clean_2009-2010.csv', index=False)
df_2011_2012.to_csv('Clean_CSVs/clean_2011-2012.csv', index=False)
df_2013_2014.to_csv('Clean_CSVs/clean_2013-2014.csv', index=False)
df_2015_2016.to_csv('Clean_CSVs/clean_2015-2016.csv', index=False)
df_2017_2018.to_csv('Clean_CSVs/clean_2017-2018.csv', index=False)
```
